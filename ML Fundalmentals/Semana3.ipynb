# Semana 3 - Primeira atividade de tranformação de dados

Our first goal is to create a column products[‘awesome’] where each row contains the number of times the word ‘awesome’ showed up in the review for the corresponding product, and 0 if the review didn’t show up.  One way to do this is to look at the each row ‘word_count’ column and follow this logic:

```python
import pandas as pd
```

```python
df =  pd.read_csv('babypro.csv')
```

```python
df
```

```python
df.info()
```

```python
def count_word_occurrences(text, word):
    if isinstance(text, str):  # Check if the 'text' is a string
        return text.lower().split().count(word.lower())
    return 0
```

```python
df['awesome'] = df['review'].apply(lambda x: count_word_occurrences(x, 'awesome') if isinstance(x, str) and 'awesome' in x.lower() else 0)
```

```python
df
```

```python
df['awesome'].value_counts()
```

```python
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

```python
for word in selected_words:
    df[word] = df['review'].apply(lambda x: count_word_occurrences(x, word) if isinstance(x, str) and word in x.lower() else 0)
```

```python
df
```

```python
word_totals = {}
for word in selected_words:
    total_occurrences = df[df[word] > 0][word].sum()
    word_totals[word] = total_occurrences
```

```python
print("Total occurrences for each word:")
print(word_totals)
```
