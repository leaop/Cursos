### **2.2. Transformação** 

Tranforme o texto da coluna `month` para sua representação numérica, exemplo: `Jan` para `1` e `Dec` para `12`.

```python
# resolução
# Criar um dicionário para mapear os meses para números
month_mapping = {
    'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4,
    'May': 5, 'Jun': 6, 'Jul': 7, 'Aug': 8,
    'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12
}

# Aplicar o mapeamento na coluna 'month'
flights['month'] = flights['month'].map(month_mapping)

# Verificar o resultado
flights.head()
```

Crie uma chave temporal `year-month` no formato `YYYY-MM` através da concatenação das colunas `year` e `month`, exemplo: `1949-01`.

```python
# resolução
# Converter 'month' para string com dois dígitos
flights['month'] = flights['month'].astype(str).str.zfill(2)

# Criar a coluna 'year-month'
flights['year-month'] = flights['year'].astype(str) + '-' + flights['month']

# Verificar o resultado
flights.head()
```

Reordene as colunas da seguinte forma: `year-month`, `year`, `month` e `passengers`.

```python
# resolução
# Reordenar as colunas
flights = flights[['year-month', 'year', 'month', 'passengers']]

# Verificar o resultado
flights.head()
```
