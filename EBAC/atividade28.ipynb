# **COVID Dashboard**

## 1\. Contexto

COVID-19 é uma doença infecciosa causada por recém-descoberto coronavírus.

Transmitido principalmente por meio de gotículas provenientes de tosses ou espirros de pessoas infectadas, a gravidade dos sintomas varia muito de pessoa para pessoa.

Fato é, não se sabe muita coisa a respeito do COVID-19. Estudos estão sendo realizados no mundo todo, porém os resultados ainda não são conclusivos e definitivos.

<center><img alt="Colaboratory logo" width="50%" src="https://raw.githubusercontent.com/carlosfab/dsnp2/master/img/covid19.jpg"></center>


Com o objetivo de elevar a consciência situacional a respeito do COVID-19, cabe uma análise de dados para revelar de forma interativa o panorama geral.

### Obtenção dos Dados

Os dados usados foram obtidos a partir do repositório, sobre a covid-19, do [OWID](https://github.com/owid/covid-19-data/tree/master/public/data) (Our World in Data).

O *dataset* utilizado para a análise será um arquivo.`csv`

* `owid-covid-data.csv` -  Dados sobre o COVID-19 (coronavírus) por Our World in Data

Neste *dataset* da COVID-19, serão apresentados dados sobre casos confirmados, mortes, hospitalizações, testes e vacinas, bem como outras variáveis de potencial interesse.

Existem diversos obstáculos na hora de analisar uma doença nova, podemos citar como exemplo, os conflitos que existem entre dados dos governo federal, estadual e dados não oficiais. Tornando assim, mais dificil a análise de uma nova doença.

* `new_deaths_smoothed_per_million` - Novas mortes atribuídas à COVID-19 (7 dias suavizados) por 1.000.000 de pessoas

* `icu_patients` - Número de pacientes COVID-19 em unidades de terapia intensiva (UTIs) em um determinado dia

* `icu_patients_per_million` - Número de pacientes COVID-19 em unidades de terapia intensiva (UTIs) em um determinado dia por 1.000.000 de pessoas

* `weekly_hosp_admissions_per_million` - Número de pacientes COVID-19 recém admitidos em hospitais em uma determinada semana por 1.000.000 de pessoas

* `total_tests` - Testes totais para a COVID-19

* `new_tests` - Novos testes para a COVID-19 (apenas calculados para dias consecutivos)

* `new_tests_smoothed_per_thousand` - Novos testes para COVID-19 (7 dias suavizados) por 1.000 pessoas

## 3\. Extração

```python
warnings.filterwarnings('ignore')

# importar o dataset COVID-19 para um DataFrame
#dataset_covid = "https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv" #direto da fonte com atulização
#df = pd.read_csv(dataset_covid)
#para apresentação
df = pd.read_csv('owid-covid-data.csv')
```

## 4\. Transformação

```python
df.dtypes
```

```python
# Obter a lista de porcentagens de valores ausentes
missing_values = (df.isnull().sum() / df.shape[0]).sort_values(ascending=False)

# Exibir a lista
print(missing_values)
```

```python
# Removendo colunas com mais de 90% de valores ausentes
threshold = 0.90
missing_fraction = df.isnull().sum() / df.shape[0]
columns_to_drop = missing_fraction[missing_fraction > threshold].index
df = df.drop(columns=columns_to_drop)
```

```python
# Preenchendo valores ausentes em colunas de séries temporais (usando interpolação)
columns_to_interpolate = ['new_cases', 'new_deaths', 'new_vaccinations']
df[columns_to_interpolate] = df[columns_to_interpolate].interpolate(method='linear', limit_direction='forward', axis=0)
```

```python
# Preenchendo valores ausentes em variáveis categóricas com o valor mais frequente
columns_to_fill_mode = ['continent', 'location']
for column in columns_to_fill_mode:
    df[column].fillna(df[column].mode()[0], inplace=True)
```

```python
# Preenchendo valores ausentes com a média para colunas numéricas relevantes
columns_to_fill_mean = ['total_cases', 'total_deaths', 'total_tests']
df[columns_to_fill_mean] = df[columns_to_fill_mean].fillna(df[columns_to_fill_mean].mean())
```

```python
# Obter a lista de porcentagens de valores ausentes
missing_values = (df.isnull().sum() / df.shape[0]).sort_values(ascending=False)

# Exibir a lista
print(missing_values)# Verificando se os Valores faltantes foram preenchidos
```

```python
# Transformar coluna "date" no formato Datetime
df.date = pd.to_datetime(df.date)
```

```python
# Criar lista com o nome dos Continentes
continentes = ['Europe', 'Asia', 'North America', 'South America', 'Oceania', 'Africa', 'European Union',]
```

## 5\. Carregamento

```python
# Localizar quais Países possuem mais casos para data mais atual '2024-04-30'
df.loc[~df.location.isin(continentes)][df.date == '2024-04-30'].sort_values(by='total_cases', ascending=False)[0:10]
```

```python
# Localizar quais Países possuem mais mortes para data mais atual '2024-04-30'
df.loc[~df.location.isin(continentes)][df.date == '2024-04-30'].sort_values(by='total_deaths', ascending=False)[0:10][['iso_code', 'continent', 'location', 'date', 'total_deaths']]
```

```python
# Localizando os dados sobre os Países
USA = df.loc[df.iso_code == 'USA']
IND = df.loc[df.iso_code == 'IND']
BRA = df.loc[df.iso_code == 'BRA']
FRA = df.loc[df.iso_code == 'FRA']
RUS = df.loc[df.iso_code == 'RUS']
MEX = df.loc[df.iso_code == 'MEX']
GBR = df.loc[df.iso_code == 'GBR']
# Localizando os dados sobre o Mundo
WRL = df.loc[df.iso_code == 'OWID_WRL']

# Criando DataFrame com os Países selecionados
paises_grafico_casos = pd.concat([USA, IND, BRA, FRA, RUS])
paises_grafico_mortes = pd.concat([USA, BRA, MEX, IND, GBR])

# Criando DataFrame com os Países selecionados e o Mundo
paisesMundo_grafico_casos = pd.concat([WRL, USA, IND, BRA, FRA, RUS])
paisesMundo_grafico_mortes = pd.concat([WRL, USA, BRA, MEX, IND, GBR])

# Localizando Variáveis na data mais recente
total_casos = paises_grafico_casos.loc[paises_grafico_casos.date == '2024-04-30',['date', 'location', 'total_cases']].sort_values(by='total_cases', ascending=False)
total_mortes = paises_grafico_mortes.loc[paises_grafico_mortes.date == '2024-04-30',['date', 'location', 'total_deaths']].sort_values(by='total_deaths', ascending=False)
total_casos_mundo = paisesMundo_grafico_casos.loc[paisesMundo_grafico_casos.date == '2024-04-30',['date', 'location', 'total_cases']].sort_values(by='total_cases', ascending=False)
total_mortes_mundo = paisesMundo_grafico_mortes.loc[paisesMundo_grafico_mortes.date == '2024-04-30',['date', 'location', 'total_deaths']].sort_values(by='total_deaths', ascending=False)


# Plotando os gráficos de barras
fig, ax = plt.subplots(nrows=1 , ncols=4, figsize=(26,8))

# Definindo estilo dos Gráficos para o estilo do Seaborn
sns.set()

# Gráfico 1
sns.barplot(x='location', y=total_casos['total_cases'], data=paises_grafico_casos, ax=ax[0], palette='rocket')
ax[0].set_title('Número de Casos Por País')
ax[0].set_ylabel('Total de Casos')
ax[0].set_xlabel('País')

# Gráfico 2
sns.barplot(x='location', y=total_casos_mundo['total_cases'], data=paisesMundo_grafico_casos, ax=ax[1], palette='rocket')
ax[1].set_title('Número de Casos por País em Comparação ao Mundo')
ax[1].set_ylabel('Total de Casos')
ax[1].set_xlabel('Localização')

# Gráfico 3
sns.barplot(x='location', y=total_mortes['total_deaths'], data=paises_grafico_mortes, ax=ax[2], palette='rocket')
ax[2].set_title('Número de Mortes por País')
ax[2].set_ylabel('Total de Mortes')
ax[2].set_xlabel('País')

# Gráfico 4
sns.barplot(x='location', y=total_mortes_mundo['total_deaths'], data=paisesMundo_grafico_mortes, ax=ax[3], palette='rocket')
ax[3].set_title('Número de Mortes por Páis em Comparação ao Mundo')
ax[3].set_ylabel('Total de Mortes')
ax[3].set_xlabel('Localização')

plt.tight_layout();

# Salvar Imagem
fig.savefig('barrasCovidMundo.png')
```
