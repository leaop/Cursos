**Dica:** Uma excelente forma de remover elementos duplicados de uma lista é convertê-la para `set` e depois para `list` novamente.



````python
# Carrega o arquivo Excel
workbook = openpyxl.load_workbook('dados_atv10.xlsx')

# Seleciona automaticamente a planilha
sheet = workbook.active

# Extrai os dados das colunas escolaridade e tipo_cartao, removendo duplicados
escolaridade = list(set(cell.value for cell in sheet['F'][1:]))
tipo_cartao = list(set(cell.value for cell in sheet['I'][1:]))

# Criando o dicionário
credito = {'tipo_cartao': tipo_cartao, 'escolaridade': escolaridade}

# Converte o dicionário para JSON
credito_json = json.dumps(credito, indent=4)

print(credito_json)
````

## 3\. BÔNUS: Texto para CSV

No arquivo de texto `ebac.txt` você encontra o texto presente no rodapé da página de cursos da EBAC ([link](https://ebaconline.com.br/)).



**Arquivo TXT:** ebac.txt

````python
%%writefile ebac.txt
MÍDIAS SOCIAIS
Instagram, Facebook, Youtube, LinkedIn 

CURSOS
Software, Design, Marketing, Audiovisual, Programação & Data, Games

WEBINARS
Próximos, Anteriores

SOBRE
Sobre nós, Centro de carreiras, Vagas

CONTATO
WhatsApp +55 (11) 4200-2991
Telefone +55 (11) 3030-3200

BLOG
Design, Audiovisual, Marketing
````

Extraia os números de contato do arquivo texto `ebac.txt` e salve-os no arquivo csv `contato_ebac.csv` com o separador `;` no seguinte formato:

```
tipo;numero
whatsapp;+551142002991
telefone;+551130303200
```

**Nota:** Escreva o código da sua solução abaixo em uma ou mais células, você não precisa enviar o arquivo csv gerado.

````python
# solução do exercício 3 (bônus)
import re
import csv


#abrir o arquivo e ler as linhas
with open(file='ebac.txt', mode='r', encoding='utf8') as arquivo:
    linhas = arquivo.readlines()

#limpar as linhas antes de qualquer coisa, já que o formato final é sem espaço
linhas = filter(lambda linha: linha != '\n', linhas) #para cada linha, verificar se não é igual a \n
linhas = map(lambda linha: linha.strip(), linhas) #extração do \n, por conta do strip aplicada em cada linha
linhas = list(linhas)

# Define padrões para encontrar números de telefone e WhatsApp
telefone_pattern = re.compile(r'\bTelefone\b[^\d]*(\+\d+)', re.IGNORECASE)
whatsapp_pattern = re.compile(r'\bWhatsApp\b[^\d]*(\+\d+)', re.IGNORECASE)

# Procura por números de telefone
telefones = [telefone_pattern.findall(linha) for linha in linhas]
telefones = [numero for sublist in telefones for numero in sublist]  # achatar a lista

# Procura por números de WhatsApp
whatsapps = [whatsapp_pattern.findall(linha) for linha in linhas]
whatsapps = [numero for sublist in whatsapps for numero in sublist]  # achatar a lista

# Cria uma lista de contatos
contatos = [
    {'tipo': 'telefone', 'numero': telefone}
    for telefone in telefones
] + [
    {'tipo': 'whatsapp', 'numero': whatsapp}
    for whatsapp in whatsapps
]

# Salva os contatos em um arquivo CSV
with open('contato_ebac.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['tipo', 'numero']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames, delimiter=';')

    writer.writeheader()
    writer.writerows(contatos)
````
